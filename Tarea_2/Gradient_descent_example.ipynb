```python
import numpy as np
from matplotlib import pylab as plt
```

```python
def fun(x, y):
    return x + y
```

```python
X, Y = np.meshgrid(np.linspace(-1,2,100),np.linspace(-1,1,100))
Z = fun(X, Y)
```

```python
ax = plt.figure().add_subplot(projection='3d')

ax.plot_surface(X, Y, Z)

ax.set_xlabel("x")
ax.set_ylabel("y")

plt.show()
```

```python
lr = 0.1
niter = 15
```

```python
init = np.array([1.5, 0.8])

approx = [init]
for iter in range(niter):
    approx.append(approx[-1] - lr * np.array([1, 1]))

approx = np.array(approx)
```

```python
fig, ax = plt.subplots()

cp = ax.contour(X, Y, Z, levels=9)
ax.clabel(cp, cp.levels)
ax.scatter(approx[:,0], approx[:,1], s=12, c=np.arange(approx.shape[0]), cmap='inferno')

plt.show()
```

1. **Para cada una de las funciones siguientes, calculen (a mano) el vector gradiente, todos los puntos críticos y el mínimo global.**

*f(x, y) = x² + y²*

**Vector Gradiente**

∂f/∂x = 2x

∂f/∂y = 2y

∇f(x, y) = [2x, 2y]

**Puntos Críticos**

∂f/∂x = 0 ⇔ 2x = 0 ⇔ x = 0

∂f/∂y = 0 ⇔ 2y = 0 ⇔ y = 0

Existe un punto crítico y es (0, 0)

**Mínimo Global**

1. *Calcular la matriz Hessiana*:

    ∂²g/∂x² = 2

    ∂²g/∂x∂y = 0

    ∂²g/∂y∂x = 0

    ∂²g/∂y² = 2

2. *Determinante*
    
    D = (∂²g/∂x² * ∂²g/∂y²) - (∂²g/∂x∂y * ∂²g/∂y∂x) 

    ⇔ D = (2 * 2) - (0 * 0)
    
    ⇔ D = 4

     Si D = 4 > 0 y ∂²g/∂x² = 2 > 0, entonces el punto (0, 0) es un mínimo global.
     
El mínimo global es el punto crítico (0, 0)

*g(x, y) = x² - y²*

**Vector Gradiente**

∂g/∂x = 2x

∂g/∂y = -2y

∇g(x, y) = [2x, -2y]

**Puntos Críticos**

∂g/∂x = 0 ⇔ 2x = 0 ⇔ x = 0

∂g/∂y = 0 ⇔ -2y = 0 ⇔ y = 0

Existe un punto crítico y es (0, 0).

**Mínimo Global**

1. *Calcular la matriz Hessiana*:

    ∂²g/∂x² = 2

    ∂²g/∂x∂y = 0

    ∂²g/∂y∂x = 0

    ∂²g/∂y² = -2

2. *Determinante*
    
    D = (∂²g/∂x² * ∂²g/∂y²) - (∂²g/∂x∂y * ∂²g/∂y∂x) 

    ⇔ D = (2 * -2) - (0 * 0)
    
    ⇔ D = -4

Por teorema, si D = -4 < 0, entonces el punto (0, 0) es un punto silla, no es un mínimo local ni un máximo local.

*h(x, y) = x⁴ - x³ - 2x² + |y|*

**Vector Gradiente**

∂h/∂x = 4x³ - 3x² - 4x

∂h/∂y = y / |y|

∇h(x, y) = [4x³ - 3x² - 4x, y / |y|]

**Puntos Críticos**

∂h/∂x = 0 ⇔ 4x³ - 3x² - 4x = 0 ⇔ x(4x² - 3x - 4) = 0

*   Caso x = 0

    x = 0

*   Caso 4x² - 3x - 4 = 0
    
    x = (-b ± √(b² - 4ac)) / 2a

    x = (3 ± √73) / 8

    x = (3 + √73) / 8  ó  x = (3 - √73) / 8

∂h/∂y = 0 ⇔ y / |y| = 0 ⇔ y = 0

Los puntos críticos son (0, 0), ((3 + √73) / 8, 0) y ((3 - √73) / 8, 0)

**Mínimo Global**

1. *Calcular la matriz Hessiana*:

    ∂²h/∂x² = 12x² - 6x - 4

    ∂²h/∂x∂y = 0

    ∂²h/∂y∂x = 0

    ∂²h/∂y² = 0

2. *Determinante*
    
    D = (∂²h/∂x² * ∂²h/∂y²) - (∂²h/∂x∂y * ∂²h/∂y∂x) 

    ⇔ D = (12x² - 6x - 4 * 0) - (0 * 0)
    
    ⇔ D = 0

Por teorema, si D = 0, el teorema no proporciona información sobre los puntos.

*h(x, y) = x⁴ - x³ - 2x² + |y|*

```python
def h(x, y):
    return x**4 - x**3 - 2 * (x**2) + abs(y)
```

```python
X, Y = np.meshgrid(np.linspace(-1.5,2,100),np.linspace(-1,1,100))
Z = h(X, Y)
```

```python
ax = plt.figure().add_subplot(projection='3d')

ax.plot_surface(X, Y, Z)

ax.set_xlabel("x")
ax.set_ylabel("y")

plt.show()
```

3. **Dibujen en otra figura varias curvas de nivel para cada función**

*f(x, y) = x² + y²*

Las curvas de nivel de *f*(x, y) son las curvas con ecuaciones f(x, y) = k donde k es una constante en el rango de f.

Esto significa que todas las curvas de nivel para *f* tienen la forma x² + y² = k, circunferencias.

Para esta función, seleccioné valores de k = 2, 4, 6 y 8.

```python
X, Y = np.meshgrid(np.linspace(-3,3,100),np.linspace(-3,3,100))
Z = f(X, Y)
```

```python
# Definir los niveles para las curvas de nivel
levels = [2, 4, 6, 8]

ax = plt.figure().add_subplot(projection='3d')

ax.plot_surface(X, Y, Z)

# Agregar las curvas de nivel en la figura tridimensional
ax.contour(X, Y, Z, levels=levels, zdir='z', offset=np.min(Z), colors='black')

# Graficar las curvas de nivel dentro de la figura tridimensional
plt.contour(X, Y, Z, levels=levels, colors='black', linewidths=2)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

*g(x, y) = x² - y²*

Las curvas de nivel de *g*(x, y) son las curvas con ecuaciones g(x, y) = k donde k es una constante en el rango de f.

Esto significa que todas las curvas de nivel para *g* tienen la forma x² - y² = k, hiperbolas.

Para esta función, seleccioné valores de k = -6, -3, 0, 3, 6.

```python
X, Y = np.meshgrid(np.linspace(-4,4,100),np.linspace(-4,4,100))
Z = g(X, Y)
```

```python
# Definir los niveles para las curvas de nivel
levels = [-6, -3, 0, 3, 6]

ax = plt.figure().add_subplot(projection='3d')

ax.plot_surface(X, Y, Z)

# Agregar las curvas de nivel en la figura tridimensional
ax.contour(X, Y, Z, levels=levels, zdir='z', offset=np.min(Z), colors='black')

# Graficar las curvas de nivel dentro de la figura tridimensional
plt.contour(X, Y, Z, levels=levels, colors='black', linewidths=2)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
