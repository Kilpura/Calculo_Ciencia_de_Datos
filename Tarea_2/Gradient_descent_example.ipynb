1. **Para cada una de las funciones siguientes, calculen (a mano) el vector gradiente, todos los puntos críticos y el mínimo global.**

*f(x, y) = x² + y²*

**Vector Gradiente**

∂f/∂x = 2x

∂f/∂y = 2y

∇f(x, y) = [2x, 2y]

**Puntos Críticos**

∂f/∂x = 0 ⇔ 2x = 0 ⇔ x = 0

∂f/∂y = 0 ⇔ 2y = 0 ⇔ y = 0

Existe un punto crítico y es (0, 0)

**Mínimo Global**

1. *Calcular la matriz Hessiana*:

    ∂²g/∂x² = 2

    ∂²g/∂x∂y = 0

    ∂²g/∂y∂x = 0

    ∂²g/∂y² = 2

2. *Determinante*
    
    D = (∂²g/∂x² * ∂²g/∂y²) - (∂²g/∂x∂y * ∂²g/∂y∂x) 

    ⇔ D = (2 * 2) - (0 * 0)
    
    ⇔ D = 4

     Si D = 4 > 0 y ∂²g/∂x² = 2 > 0, entonces el punto (0, 0) es un mínimo global.
     
El mínimo global es el punto crítico (0, 0)

*g(x, y) = x² - y²*

**Vector Gradiente**

∂g/∂x = 2x

∂g/∂y = -2y

∇g(x, y) = [2x, -2y]

**Puntos Críticos**

∂g/∂x = 0 ⇔ 2x = 0 ⇔ x = 0

∂g/∂y = 0 ⇔ -2y = 0 ⇔ y = 0

Existe un punto crítico y es (0, 0).

**Mínimo Global**

1. *Calcular la matriz Hessiana*:

    ∂²g/∂x² = 2

    ∂²g/∂x∂y = 0

    ∂²g/∂y∂x = 0

    ∂²g/∂y² = -2

2. *Determinante*
    
    D = (∂²g/∂x² * ∂²g/∂y²) - (∂²g/∂x∂y * ∂²g/∂y∂x) 

    ⇔ D = (2 * -2) - (0 * 0)
    
    ⇔ D = -4

Por teorema, si D = -4 < 0, entonces el punto (0, 0) es un punto silla, no es un mínimo local ni un máximo local.

*h(x, y) = x⁴ - x³ - 2x² + |y|*

**Vector Gradiente**

∂h/∂x = 4x³ - 3x² - 4x

∂h/∂y = y / |y|

∇h(x, y) = [4x³ - 3x² - 4x, y / |y|]

**Puntos Críticos**

∂h/∂x = 0 ⇔ 4x³ - 3x² - 4x = 0 ⇔ x(4x² - 3x - 4) = 0

*   Caso x = 0

    x = 0

*   Caso 4x² - 3x - 4 = 0
    
    x = (-b ± √(b² - 4ac)) / 2a

    x = (3 ± √73) / 8

    x = (3 + √73) / 8  ó  x = (3 - √73) / 8

∂h/∂y = 0 ⇔ y / |y| = 0 ⇔ y = 0

Los puntos críticos son (0, 0), ((3 + √73) / 8, 0) y ((3 - √73) / 8, 0)

**Mínimo Global**

1. *Calcular la matriz Hessiana*:

    ∂²h/∂x² = 12x² - 6x - 4

    ∂²h/∂x∂y = 0

    ∂²h/∂y∂x = 0

    ∂²h/∂y² = 0

2. *Determinante*
    
    D = (∂²h/∂x² * ∂²h/∂y²) - (∂²h/∂x∂y * ∂²h/∂y∂x) 

    ⇔ D = (12x² - 6x - 4 * 0) - (0 * 0)
    
    ⇔ D = 0

Por teorema, si D = 0, el teorema no proporciona información sobre los puntos.

5. Implementen el algoritmo de descenso de gradiente.

```python
def descenso_gradiente(funcion, gradiente, tasa_aprendizaje, x_inicial, iteraciones):
     x = x_inicial
     approx = [x_inicial]

     for iter in range(iteraciones):
          grad = gradiente(x)
          nueva_x = x - tasa_aprendizaje * grad
          approx.append(nueva_x)
     
          x = nueva_x
     
     return np.array(approx)
```

```python
def figura_gradiente(funcion, eje_x, eje_y, approx):
     X, Y = np.meshgrid(np.linspace(*eje_x, 100),np.linspace(*eje_y, 100))
     Z = funcion([X, Y])

     fig, ax = plt.subplots()
     cp = ax.contour(X, Y, Z, levels=9)
     ax.clabel(cp, cp.levels)
     ax.scatter(approx[:,0], approx[:,1], s=12, c=np.arange(approx.shape[0]), cmap='inferno')
     plt.show()
```

```python
# Definir las funciones de perdida
def f(x):
     return x[0]**2 + x[1]**2

def g(x):
    return x[0]**2 - x[1]**2

def h(x):
    return x[0]**4 - x[0]**3 - 2*x[0]**2 + np.abs(x[1])
```

```python
# Gradientes
def gradiente_f(x):
     return np.array([2*x[0], 2*x[1]])

def gradiente_g(x):
    return np.array([2*x[0], -2*x[1]])

def gradiente_h(x):
    grad_y = 0 if x[1] == 0 else x[1] / np.abs(x[1])
    return np.array([4*x[0]**3 - 3*x[0]**2 - 4*x[0], grad_y])
```
