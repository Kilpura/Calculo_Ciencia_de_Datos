```python
import numpy as np
from matplotlib import pylab as plt
```

```python
def fun(x, y):
    return x + y
```

```python
X, Y = np.meshgrid(np.linspace(-1,2,100),np.linspace(-1,1,100))
Z = fun(X, Y)
```

```python
ax = plt.figure().add_subplot(projection='3d')

ax.plot_surface(X, Y, Z)

ax.set_xlabel("x")
ax.set_ylabel("y")

plt.show()
```

```python
lr = 0.1
niter = 15
```

```python
init = np.array([1.5, 0.8])

approx = [init]
for iter in range(niter):
    approx.append(approx[-1] - lr * np.array([1, 1]))

approx = np.array(approx)
```

```python
fig, ax = plt.subplots()

cp = ax.contour(X, Y, Z, levels=9)
ax.clabel(cp, cp.levels)
ax.scatter(approx[:,0], approx[:,1], s=12, c=np.arange(approx.shape[0]), cmap='inferno')

plt.show()
```

*f(x, y) = x + y*

**Vector Gradiente**

f/x = 2x

f/y = 2y

f(x, y) = [2x, 2y]

**Puntos Crticos**

f/x = 0  2x = 0  x = 0

f/y = 0  2y = 0  y = 0

Existe un punto crtico y es (0, 0)

**Mnimo Global**

1. *Calcular la matriz Hessiana*:

    g/x = 2

    g/xy = 0

    g/yx = 0

    g/y = 2

2. *Determinante*
    
    D = (g/x * g/y) - (g/xy * g/yx) 

     D = (2 * 2) - (0 * 0)
    
     D = 4

     Si D = 4 > 0 y g/x = 2 > 0, entonces el punto (0, 0) es un mnimo global.
     
El mnimo global es el punto crtico (0, 0)

*g(x, y) = x - y*

**Vector Gradiente**

g/x = 2x

g/y = -2y

g(x, y) = [2x, -2y]

**Puntos Crticos**

g/x = 0  2x = 0  x = 0

g/y = 0  -2y = 0  y = 0

Existe un punto crtico y es (0, 0).

**Mnimo Global**

1. *Calcular la matriz Hessiana*:

    g/x = 2

    g/xy = 0

    g/yx = 0

    g/y = -2

2. *Determinante*
    
    D = (g/x * g/y) - (g/xy * g/yx) 

     D = (2 * -2) - (0 * 0)
    
     D = -4

Por teorema, si D = -4 < 0, entonces el punto (0, 0) es un punto silla, no es un mnimo local ni un mximo local.

3. **Dibujen en otra figura varias curvas de nivel para cada funcin**

*f(x, y) = x + y*

Las curvas de nivel de *f*(x, y) son las curvas con ecuaciones f(x, y) = k donde k es una constante en el rango de f.

Esto significa que todas las curvas de nivel para *f* tienen la forma x + y = k, circunferencias.

Para esta funcin, seleccion valores de k = 2, 4, 6 y 8.

```python
X, Y = np.meshgrid(np.linspace(-3,3,100),np.linspace(-3,3,100))
Z = f(X, Y)
```

```python
# Definir los niveles para las curvas de nivel
levels = [2, 4, 6, 8]

ax = plt.figure().add_subplot(projection='3d')

ax.plot_surface(X, Y, Z)

# Agregar las curvas de nivel en la figura tridimensional
ax.contour(X, Y, Z, levels=levels, zdir='z', offset=np.min(Z), colors='black')

# Graficar las curvas de nivel dentro de la figura tridimensional
plt.contour(X, Y, Z, levels=levels, colors='black', linewidths=2)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

*g(x, y) = x - y*

Las curvas de nivel de *g*(x, y) son las curvas con ecuaciones g(x, y) = k donde k es una constante en el rango de f.

Esto significa que todas las curvas de nivel para *g* tienen la forma x - y = k, hiperbolas.

Para esta funcin, seleccion valores de k = -6, -3, 0, 3, 6.

```python
X, Y = np.meshgrid(np.linspace(-4,4,100),np.linspace(-4,4,100))
Z = g(X, Y)
```

```python
# Definir los niveles para las curvas de nivel
levels = [-6, -3, 0, 3, 6]

ax = plt.figure().add_subplot(projection='3d')

ax.plot_surface(X, Y, Z)

# Agregar las curvas de nivel en la figura tridimensional
ax.contour(X, Y, Z, levels=levels, zdir='z', offset=np.min(Z), colors='black')

# Graficar las curvas de nivel dentro de la figura tridimensional
plt.contour(X, Y, Z, levels=levels, colors='black', linewidths=2)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

4. Dibujen todos los puntos crticos que calcularon para cada funcin como un marker en la figura anterior.

*f(x, y) = x + y*

Habamos obtenido que el nico punto crtico es (0, 0).

```python
X, Y = np.meshgrid(np.linspace(-3,3,100),np.linspace(-3,3,100))
Z = f(X, Y)
```
