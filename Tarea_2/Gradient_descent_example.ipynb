```python
def fun(x, y):
    return x + y
```

*f(x, y) = x + y*

**Vector Gradiente**

f/x = 2x

f/y = 2y

f(x, y) = [2x, 2y]

**Puntos Crticos**

f/x = 0  2x = 0  x = 0

f/y = 0  2y = 0  y = 0

Existe un punto crtico y es (0, 0)

**Mnimo Global**

1. *Calcular la matriz Hessiana*:

    g/x = 2

    g/xy = 0

    g/yx = 0

    g/y = 2

2. *Determinante*
    
    D = (g/x * g/y) - (g/xy * g/yx) 

     D = (2 * 2) - (0 * 0)
    
     D = 4

     Si D = 4 > 0 y g/x = 2 > 0, entonces el punto (0, 0) es un mnimo global.
     
El mnimo global es el punto crtico (0, 0)

*g(x, y) = x - y*

**Vector Gradiente**

g/x = 2x

g/y = -2y

g(x, y) = [2x, -2y]

**Puntos Crticos**

g/x = 0  2x = 0  x = 0

g/y = 0  -2y = 0  y = 0

Existe un punto crtico y es (0, 0).

**Mnimo Global**

1. *Calcular la matriz Hessiana*:

    g/x = 2

    g/xy = 0

    g/yx = 0

    g/y = -2

2. *Determinante*
    
    D = (g/x * g/y) - (g/xy * g/yx) 

     D = (2 * -2) - (0 * 0)
    
     D = -4

Por teorema, si D = -4 < 0, entonces el punto (0, 0) es un punto silla, no es un mnimo local ni un mximo local.

5. Implementen el algoritmo de descenso de gradiente.

```python
def descenso_gradiente(funcion, gradiente, tasa_aprendizaje, x_inicial, iteraciones):
     x = x_inicial
     approx = [x_inicial]

     for iter in range(iteraciones):
          grad = gradiente(x)
          nueva_x = x - tasa_aprendizaje * grad
          approx.append(nueva_x)
     
          x = nueva_x
     
     return np.array(approx)
```

```python
def figura_gradiente(funcion, eje_x, eje_y, approx):
     X, Y = np.meshgrid(np.linspace(*eje_x, 100),np.linspace(*eje_y, 100))
     Z = funcion([X, Y])

     fig, ax = plt.subplots()
     cp = ax.contour(X, Y, Z, levels=9)
     ax.clabel(cp, cp.levels)
     ax.scatter(approx[:,0], approx[:,1], s=12, c=np.arange(approx.shape[0]), cmap='inferno')
     plt.show()
```

```python
# Definir las funciones de perdida
def f(x):
     return x[0]**2 + x[1]**2

def g(x):
    return x[0]**2 - x[1]**2

def h(x):
    return x[0]**4 - x[0]**3 - 2*x[0]**2 + np.abs(x[1])
```

```python
# Gradientes
def gradiente_f(x):
     return np.array([2*x[0], 2*x[1]])

def gradiente_g(x):
    return np.array([2*x[0], -2*x[1]])

def gradiente_h(x):
    grad_y = 0 if x[1] == 0 else x[1] / np.abs(x[1])
    return np.array([4*x[0]**3 - 3*x[0]**2 - 4*x[0], grad_y])
```
