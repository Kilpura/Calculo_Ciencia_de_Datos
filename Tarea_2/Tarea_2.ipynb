# 1.
Para cada una de las funciones calculen a mano el vector gradiente, todos los puntos críticos y el mínimo global.

```python
def f(x, y):
     return x**2 + y**2
```

**Vector Gradiente**

∂f/∂x = 2x

∂f/∂y = 2y

∇f(x, y) = [2x, 2y]

```python
def f_gradiente(x, y):
     return [2*x, 2*y]
```

**Punto Crítico**

∂f/∂x = 0 ⇔ 2x = 0 ⇔ x = 0

∂f/∂y = 0 ⇔ 2y = 0 ⇔ y = 0

Existe un punto crítico y es (0, 0)

**Mínimo Global**

1. *Calcular la matriz Hessiana*:

    ∂²g/∂x² = 2

    ∂²g/∂x∂y = 0

    ∂²g/∂y∂x = 0

    ∂²g/∂y² = 2

2. *Determinante*
    
    D = (∂²g/∂x² * ∂²g/∂y²) - (∂²g/∂x∂y * ∂²g/∂y∂x) 

    ⇔ D = (2 * 2) - (0 * 0)
    
    ⇔ D = 4

     Si D = 4 > 0 y ∂²g/∂x² = 2 > 0, entonces el punto (0, 0) es un mínimo global.


```python
def g(x, y):
    return x**2 - y**2
```

**Vector Gradiente**

∂g/∂x = 2x

∂g/∂y = -2y

∇g(x, y) = [2x, -2y]


```python
def g_gradiente(x, y):
    return [2*x, -2*y]
```

**Puntos Críticos**

∂g/∂x = 0 ⇔ 2x = 0 ⇔ x = 0

∂g/∂y = 0 ⇔ -2y = 0 ⇔ y = 0

Existe un punto crítico y es (0, 0).

**Mínimo Global**

1. *Calcular la matriz Hessiana*:

    ∂²g/∂x² = 2

    ∂²g/∂x∂y = 0

    ∂²g/∂y∂x = 0

    ∂²g/∂y² = -2

2. *Determinante*
    
    D = (∂²g/∂x² * ∂²g/∂y²) - (∂²g/∂x∂y * ∂²g/∂y∂x) 

    ⇔ D = (2 * -2) - (0 * 0)
    
    ⇔ D = -4

Por teorema, si D = -4 < 0, entonces el punto (0, 0) es un punto silla, no es un mínimo local ni un máximo local.

```python
def h(x, y):
    return x**4 - x**3 - 2*x**2 + abs(y)
```

**Vector Gradiente**

∂h/∂x = 4x³ - 3x² - 4x

∂h/∂y = y / |y|

∇h(x, y) = [4x³ - 3x² - 4x, y / |y|]

```python
def h_gradiente(x, y):
    grad_y = 0 if y == 0 else y / abs(y)
    return [4*x**3 - 3*x**2 - 4*x, grad_y]
```

**Puntos Críticos**

∂h/∂x = 0 ⇔ 4x³ - 3x² - 4x = 0 ⇔ x(4x² - 3x - 4) = 0

*   Caso x = 0

    x = 0

*   Caso 4x² - 3x - 4 = 0
    
    x = (-b ± √(b² - 4ac)) / 2a

    x = (3 ± √73) / 8

    x = 1.443  ó  x = -0.693

∂h/∂y = 0 ⇔ y / |y| = 0

El único valor que resuelve esta expresión es y = 0, no obstante no es concluyente el punto crítico por al división por cero.

Los posibles puntos críticos son  (0, 0), (1.443, 0) y (-0.693, 0)

**Mínimo Global**

1. *Calcular la matriz Hessiana*:

    ∂²h/∂x² = 12x² - 6x - 4

    ∂²h/∂x∂y = 0

    ∂²h/∂y∂x = 0

    ∂²h/∂y² = 0

2. *Determinante*
    
    D = (∂²h/∂x² * ∂²h/∂y²) - (∂²h/∂x∂y * ∂²h/∂y∂x) 

    ⇔ D = (12x² - 6x - 4 * 0) - (0 * 0)
    
    ⇔ D = 0

Por teorema, si D = 0, el teorema no proporciona información sobre los puntos.

# 5
Implementen el algoritmo de descenso de gradiente

```python
def descenso_gradiente(funcion, gradiente, tasa_aprendizaje, x_inicial, iteraciones):
     x = np.array(x_inicial)
     approx = [x_inicial]

     for iter in range(iteraciones):
          grad = np.array(gradiente(x[0], x[1]))
          nueva_x = x - tasa_aprendizaje * grad
          approx.append(nueva_x)
     
          x = nueva_x
     
     return np.array(approx)
```

```python
def figura_gradiente(funcion, eje_x, eje_y, approx):
     X, Y = np.meshgrid(np.linspace(*eje_x, 100),np.linspace(*eje_y, 100))
     Z = funcion(X, Y)

     fig, ax = plt.subplots()
     cp = ax.contour(X, Y, Z, levels=9)
     ax.clabel(cp, cp.levels)
     ax.scatter(approx[:,0], approx[:,1], s=12, c=np.arange(approx.shape[0]), cmap='inferno')
     plt.show()
```

Se observa que si se disminuye la tasa de iteraciones, en un principio el algoritmo de descenso de gradiente converge a un punto entre -1 y 0. Por otro lado, al aumentar el número de iteraciones, el algoritmo tiene más oportunidades de explorar diferentes áreas del espacio. Esto lleva a una convergencia en regiones que pueden estar más lejos del punto inicial y que podrían corresponder a mínimos locales más profundos o incluso el mínimo global, en este caso, las figuras muestran una convergencia entre 4 y 8.

Al ajustar la tasa de aprendizaje en pequeños incrementos, (0.014, 0.015, 0.016), es posible notar una leve diferencia en el comportamiento del algoritmo. Esta tasa de aprendizaje pequeña permite dar "pasos" más pequeños y notar una convergencia mas estable.
